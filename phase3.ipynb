### Advance opration with business example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data strucher: [restaurant_id,2021 ,2022 ,2023 ,2024]
sales_data=np.array([
    [1, 15000, 18000 ,22000 ,25000],    #biryani
    [2, 12000, 14000, 16000, 19000],    # beijing bites
    [3, 20000, 23000, 26000, 30000],     #pizza hub
    [4, 18000, 21000, 24000, 27000],     #burger point
    [5, 16000, 18000, 20500, 23000]       # chai point 
])

print("=== zomato sales analysis ===")
print("\n sales data shape", sales_data.shape)
print("\n sample data for 1st 3 restaurent",sales_data[:,1:])
```

```python
# totel sales per year
print(np.sum(sales_data,axis=0))
years_total =np.sum(sales_data[:,1:],axis=0)           #axis=0(row),1=culumb start
print("oll over year sum",years_total)

# Minimam sale per restaurant
min_sales=np.min(sales_data[:,1:],axis=1)               #axis=1(culmun) ,1=culumb start
print(min_sales)

#Maximum sale per year
max_sale=np.max(sales_data[:,1:],axis=0)
print(max_sale)\
    
#Avrerage sales per restautant
avg_sale=np.mean(sales_data[:,1:] , axis=1)
print(avg_sale)    
```

```python
cumsum=np.cumsum(sales_data[:,1:],axis=0)
print(cumsum)

plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Avrage cumulativ sales accross all restaurant")
plt.xlabel("Years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

### Vaters 

```python
vector1=np.array([1,2,3,4])
vector2=np.array([5,6,7,8])

print("\n Vector addition",vector1+vector2)

print("\n Vector multiplication",vector1*vector2)

print("\n Dot product vector",np.dot(vector1,vector2))     # v1*v2=sum

angle=np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2))) # formula angle vector
print(angle)
```
